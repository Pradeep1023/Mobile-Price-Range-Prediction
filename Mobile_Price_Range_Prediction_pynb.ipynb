```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
# Taking dependent and independent variables
X = mobile_df.drop(labels='price_range', axis=1)
Y = mobile_df['price_range']
```

```python
# Training the dataset
x_train, x_test, y_train, y_test = train_test_split(X , Y, test_size=0.2, random_state = 1)
```

```python
# transforming the data 
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# **Implementing LogisticRegression**

```python
log = LogisticRegression(fit_intercept=True, max_iter=10000)
log.fit(x_train,y_train)
```

```python
log.coef_
```

```python
log.intercept_
```

```python
train_log = log.predict(x_train)
test_log = log.predict(x_test)
```

```python
cm_train = confusion_matrix(y_train,train_log)
print(cm_train)
ax = plt.subplot()
sns.heatmap(cm_train,annot=True,ax=ax)
ax.set_xlabel('Predicted Range')
ax.set_ylabel('True Range')
ax.set_title('Confusion Matrix')
```

```python
cm_test = confusion_matrix(y_test,test_log)
print(cm_test)
ax = plt.subplot()
sns.heatmap(cm_test,annot=True,ax=ax)
ax.set_xlabel('Predicted Range')
ax.set_ylabel('True Range')
ax.set_title('Confusion Matrix')
```

```python
accuracy_score(y_train,train_log)
```

```python
Log = accuracy_score(y_test,test_log)
Log
```

```python
print(classification_report(y_train,train_log))
```

```python
# 87 % accuracy with logistic regression
```

# **Fitting Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
train_dc = dt.predict(x_train)
test_dc = dt.predict(x_test)
```

```python
cm_dc_test = confusion_matrix(y_test,test_dc)
print(cm_dc_test)
ax = plt.subplot()
sns.heatmap(cm_dc_test,annot=True,ax=ax)
ax.set_xlabel('Precited Range')
ax.set_ylabel('True Range')
ax.set_title('Confusion Matrix')
```

```python
DT = accuracy_score(y_test,test_dc)
DT
```

```python
# 84% accuracy score with Decision Tree
```

```python
print(classification_report(y_test,test_dc))
```

```python
# decision tree with cross-validation
from sklearn.model_selection import GridSearchCV
param_grid = {'max_depth':(5,20),'max_leaf_nodes':(10,80)}
dt_cv = GridSearchCV(dt, param_grid=param_grid, scoring='accuracy',cv=10, verbose=20)
dt_cv.fit(x_train,y_train)
```

```python
train_dtcv = dt_cv.predict(x_train) 
test_dtcv = dt_cv.predict(x_test)
```

```python
accuracy_score(y_train,train_dtcv)
```

```python
DT_CV = accuracy_score(y_test,test_dtcv)
DT_CV
```

```python
# 85% accuracy with cross-validation
```

```python
print(classification_report(y_test,test_dtcv))
```

# **Fitting Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
n_estimators = [50,80,100]
max_depth = [4,6,8]
min_samples_split = [50,100,150]
min_samples_leaf = [40,50]
param_dict = {'n_estimators':n_estimators,
              'max_depth':max_depth,
              'min_samples_split':min_samples_split,
              'min_samples_leaf':min_samples_leaf}
```

```python
rf_model = RandomForestClassifier()
rf_grid = GridSearchCV(estimator=rf_model,param_grid=param_dict,cv=5,verbose=2)
```

```python
rf_grid.fit(x_train,y_train)
```

```python
rf_optimal_model = rf_grid.best_estimator_
rf_optimal_model
```

```python
train_rf = rf_optimal_model.predict(x_train)
test_rf = rf_optimal_model.predict(x_test)
```

```python
accuracy_score(y_train,train_rf)
```

```python
RF = accuracy_score(y_test,test_rf)
RF
```

```python
# 81% accuracy with Random Forest
```

```python
print(classification_report(y_test,test_rf))
```

```python
# Plotting graph for feature importances
features = X.columns
importances = rf_optimal_model.feature_importances_
indices = np.argsort(importances)
```

```python
plt.figure(figsize=(10,8))
plt.title('Feature Importances')
plt.barh(range(len(indices)),importances[indices],color ='Red',align='center')
plt.yticks(range(len(indices)),[features[i] for i in indices])
plt.xlabel('Relative_importance')
plt.show() 
```
